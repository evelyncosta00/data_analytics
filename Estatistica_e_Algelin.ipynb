```python
import matplotlib.pyplot as plt
```

```python
x1,y1 = -1,0
x2,y2 = 0,1
x3,y3 = 1,2
x4,y4 = 2,1

x= [x1,x2,x3,x4]
y= [ y1,y2,y3,y4]


plt.figure(figsize = (10,5))

plt.scatter(x1,y1, color = 'green', label = 'Pares ordenados (x,y)')
plt.scatter(x2,y2, color = 'green')
plt.scatter(x3,y3, color = 'green')
plt.scatter(x4,y4, color = 'green')

plt.plot (x,y, label ='Gráfico que liga os pontos')
plt.grid(True)
plt.box(False)
plt.legend()
plt.title("Interpolação da reta sobre todos os pontos")
```

```python
# y=mx=n
# 1 = 0*m + 1*n
A = np.array([[-1,1],[0,1],[1,1],[2,1]])
y = np.array([[y1],[y2],[y3],[y4]])
```

```python
A
```

```python
y
```

```python
#X=(ATA)−1ATY --> método dos minimos quadrados MMQ

(m,n) = np.dot(np.dot(np.linalg.pinv(np.dot(A.T,A)),A.T),y)
print(m,n)
```

```python
x1,y1 = -1,0
x2,y2 = 0,1
x3,y3 = 1,2
x4,y4 = 2,1

x= [x1,x2,x3,x4]
y= [ y1,y2,y3,y4]


plt.figure(figsize = (10,5))

plt.scatter(x1,y1, color = 'green', label = 'Pares ordenados (x,y)')
plt.scatter(x2,y2, color = 'green')
plt.scatter(x3,y3, color = 'green')
plt.scatter(x4,y4, color = 'green')

v_x = np.linspace(-1,2)
plt.plot (x,y, label ='Gráfico que liga os pontos')
plt.plot(v_x, m*v_x + n, label = 'Reta otimizada pelo MMQ' )
plt.grid(True)
plt.box(False)
plt.legend()
plt.title("Interpolação da reta sobre todos os pontos")
```
